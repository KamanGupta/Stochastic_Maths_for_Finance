```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy.stats import norm

np.random.seed(42)

T, n = 1.0, 10000
dt = T / n
t = np.linspace(0, T, n + 1)

dW = np.random.normal(0, np.sqrt(dt), n)
W = np.concatenate([[0], np.cumsum(dW)])

qv = np.sum(dW**2)
tv = np.sum(np.abs(dW))

print("Part (e)")
print(f"Empirical QV : {qv:.6f}")
print(f"Theoretical QV : {T:.6f}")
print(f"Total Variation : {tv:.4f}")

n_vals = [10, 50, 100, 500, 1000, 5000, 10000]
qv_vals = []

for m in n_vals:
    dt_m = T / m
    dW_m = np.random.normal(0, np.sqrt(dt_m), m)
    qv_vals.append(np.sum(dW_m**2))

t0 = 0.5
m = 10000
Wt = np.random.normal(0, np.sqrt(t0), m)

print("\nPart (g)")
print(f"Mean : {np.mean(Wt):.6f}")
print(f"Variance : {np.var(Wt):.6f}")

fig = plt.figure(figsize=(16, 12))
gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.4, wspace=0.35)

ax1 = fig.add_subplot(gs[0, :])
ax1.plot(t, W, lw=0.6)
ax1.axhline(0, color='gray', ls='--')
ax1.set_title(f"Single Brownian Path (N={n})\nQV={qv:.4f}, TV={tv:.2f}")
ax1.set_xlabel("t")
ax1.set_ylabel("W(t)")

ax2 = fig.add_subplot(gs[1, 0])
ax2.semilogx(n_vals, qv_vals, 'o-')
ax2.axhline(T, color='red', ls='--', label='T=1')
ax2.set_title("QV Convergence")
ax2.set_xlabel("N")
ax2.set_ylabel("QV")
ax2.legend()
ax2.grid(alpha=0.3)

ax3 = fig.add_subplot(gs[1, 1])
x = np.linspace(-3, 3, 300)
ax3.hist(Wt, bins=60, density=True, alpha=0.6, label='Simulated')
ax3.plot(x, norm.pdf(x, 0, np.sqrt(t0)), 'r-', lw=2,
         label='N(0,0.5)')
ax3.set_title("Distribution of $W_{0.5}$")
ax3.set_xlabel("$W(0.5)$")
ax3.set_ylabel("Density")
ax3.legend()

plt.savefig("q1_plot.png", dpi=150, bbox_inches="tight")
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy.stats import lognorm, norm

np.random.seed(42)

S0, mu, sig = 100.0, 0.07, 0.2
T, n = 1.0, 500
dt = T / n
m = 50000

t_err = 10

corr = S0 * np.exp(mu * t_err)
incorr = S0 * np.exp((mu + sig**2 / 2) * t_err)

print("Part (c)")
print(f"Correct E[S_10]   = {corr:.4f}")
print(f"Incorrect E[S_10] = {incorr:.4f}")
print(f"Relative Error    = {(incorr/corr - 1)*100:.2f}%")

dW = np.random.normal(0, np.sqrt(dt), size=(m, n))

S = np.zeros((m, n + 1))
S[:, 0] = S0

for k in range(n):
    S[:, k+1] = (
        S[:, k]
        + mu * S[:, k] * dt
        + sig * S[:, k] * dW[:, k]
    )

ST = S[:, -1]

mu_ln = np.log(S0) + (mu - 0.5 * sig**2) * T
sd_ln = sig * np.sqrt(T)

est_mean = S0 * np.exp(mu * T)
est_std = S0 * np.exp(mu * T) * np.sqrt(np.exp(sig**2 * T) - 1)

print("\nPart (e)")
print(f"Empirical Mean = {np.mean(ST):.4f}")
print(f"Theoretical Mean = {est_mean:.4f}")
print(f"Empirical Std = {np.std(ST):.4f}")
print(f"Theoretical Std = {est_std:.4f}")

K = 110.0
sigs = [0.1, 0.3, 0.5]

res = {}

print("\nPart (f)")
print(f"{'sigma':<10}{'E[ST]':<15}{'Std[ST]':<15}{'Call Payoff':<15}")

for s in sigs:

    dW_s = np.random.normal(0, np.sqrt(dt), size=(m, n))

    P = np.zeros((m, n + 1))
    P[:, 0] = S0

    for k in range(n):
        P[:, k+1] = (
            P[:, k]
            + mu * P[:, k] * dt
            + s * P[:, k] * dW_s[:, k]
        )

    ST_s = P[:, -1]

    e = np.mean(ST_s)
    sd = np.std(ST_s)
    call = np.mean(np.maximum(ST_s - K, 0))

    res[s] = ST_s

    print(f"{s:<10.1f}{e:<15.4f}{sd:<15.4f}{call:<15.4f}")

lr = np.log(ST / S0)

th_mean = (mu - 0.5 * sig**2) * T
th_var = sig**2 * T

print("\nPart (g)")
print(f"Empirical Mean = {np.mean(lr):.6f}")
print(f"Theoretical Mean = {th_mean:.6f}")
print(f"Empirical Variance = {np.var(lr):.6f}")
print(f"Theoretical Variance = {th_var:.6f}")

fig = plt.figure(figsize=(16, 13))
gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.4, wspace=0.35)

ax1 = fig.add_subplot(gs[0, 0])

x = np.linspace(ST.min(), min(ST.max(), 400), 500)
pdf = lognorm.pdf(x, s=sd_ln, scale=np.exp(mu_ln))

ax1.hist(ST, bins=100, density=True, alpha=0.55, label='Simulated')
ax1.plot(x, pdf, 'r-', lw=2, label='Lognormal PDF')

ax1.set_title("Terminal Price Distribution")
ax1.set_xlabel("$S_T$")
ax1.set_ylabel("Density")
ax1.legend()

ax2 = fig.add_subplot(gs[0, 1])

for s in sigs:
    ax2.hist(
        res[s],
        bins=80,
        density=True,
        alpha=0.45,
        label=f"$\\sigma={s}$"
    )

ax2.axvline(K, color='red', ls='--', label=f'K={K}')
ax2.set_title("Different Volatility Levels")
ax2.set_xlabel("$S_T$")
ax2.set_ylabel("Density")
ax2.legend()

ax3 = fig.add_subplot(gs[1, 0])

x_lr = np.linspace(lr.min(), lr.max(), 400)

ax3.hist(lr, bins=80, density=True,
         alpha=0.55,
         label='Simulated')

ax3.plot(
    x_lr,
    norm.pdf(x_lr, th_mean, np.sqrt(th_var)),
    'r-',
    lw=2,
    label='Normal PDF'
)

ax3.set_title("Log-Return Distribution")
ax3.set_xlabel(r"$\ln(S_T/S_0)$")
ax3.set_ylabel("Density")
ax3.legend()

ax4 = fig.add_subplot(gs[1, 1])

t = np.linspace(0, T, n + 1)

for i in range(30):
    ax4.plot(t, S[i], lw=0.6, alpha=0.6)

ax4.set_title("Sample GBM Paths")
ax4.set_xlabel("t")
ax4.set_ylabel("$S_t$")

plt.savefig("q2_plot.png", dpi=150, bbox_inches="tight")
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy.stats import norm

np.random.seed(42)

th, sig, x0 = 2.0, 0.5, 3.0

T1, T2 = 5.0, 10.0
n1, n2 = 5000, 10000

dt1 = T1 / n1
dt2 = T2 / n2

var = sig**2 / (2 * th)
sd = np.sqrt(var)

t1 = np.linspace(0, T1, n1 + 1)
t2 = np.linspace(0, T2, n2 + 1)

print("Part (b)")
print(f"2/theta = {2/th:.4f}")
print(f"dt = {dt1:.6f}")

print("\nPart (c)")
print(f"Variance = {var:.6f}")
print(f"Std Dev = {sd:.6f}")

m = 10

p1 = np.zeros((m, n1 + 1))
p1[:, 0] = x0

for k in range(n1):
    dW = np.random.normal(0, np.sqrt(dt1), m)
    p1[:, k+1] = p1[:, k] - th * p1[:, k] * dt1 + sig * dW

m2 = 20000

p2 = np.zeros((m2, n2 + 1))
p2[:, 0] = x0

for k in range(n2):
    dW = np.random.normal(0, np.sqrt(dt2), m2)
    p2[:, k+1] = (
        p2[:, k]
        - th * p2[:, k] * dt2
        + sig * dW
    )

XT = p2[:, -1]

print("\nPart (e)")
print(f"Empirical Mean = {np.mean(XT):.6f}")
print(f"Empirical Variance = {np.var(XT):.6f}")

ths = [0.5, 2.0, 5.0]
res = {}

print("\nPart (f)")

for t in ths:

    P = np.zeros((m, n1 + 1))
    P[:, 0] = x0

    for k in range(n1):
        dW = np.random.normal(0, np.sqrt(dt1), m)
        P[:, k+1] = P[:, k] - t * P[:, k] * dt1 + sig * dW

    res[t] = P

    s = np.sqrt(sig**2 / (2 * t))
    print(f"theta={t}, sigma_inf={s:.4f}")

fig = plt.figure(figsize=(16, 14))
gs = gridspec.GridSpec(3, 2, figure=fig, hspace=0.5, wspace=0.35)

ax1 = fig.add_subplot(gs[0, :])

for i in range(m):
    ax1.plot(t1, p1[i], lw=0.8, alpha=0.7)

ax1.axhline(0, color='black', ls='--')
ax1.fill_between(t1, -2*sd, 2*sd, alpha=0.12)

ax1.set_title("OU Paths")
ax1.set_xlabel("t")
ax1.set_ylabel("X(t)")

ax2 = fig.add_subplot(gs[1, 0])

x = np.linspace(-1.5, 1.5, 400)

ax2.hist(
    XT,
    bins=80,
    density=True,
    alpha=0.55,
    label='Simulated'
)

ax2.plot(
    x,
    norm.pdf(x, 0, sd),
    'r-',
    lw=2,
    label='Normal PDF'
)

ax2.set_title("Stationary Distribution")
ax2.set_xlabel("$X_T$")
ax2.set_ylabel("Density")
ax2.legend()

for idx, t in enumerate(ths):

    ax = fig.add_subplot(gs[1 + idx // 2,
                            idx % 2 + (1 if idx == 2 else 0)])

    s = np.sqrt(sig**2 / (2 * t))

    for i in range(m):
        ax.plot(t1, res[t][i], lw=0.7, alpha=0.7)

    ax.axhline(0, color='black', ls='--')
    ax.fill_between(t1, -2*s, 2*s, alpha=0.12)

    ax.set_title(f"$\\theta={t}$")
    ax.set_xlabel("t")
    ax.set_ylabel("X(t)")

plt.savefig("q3_plot.png", dpi=150, bbox_inches="tight")
plt.close()
```
