```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)

nv = [10, 50, 200]

m = 10000

fig, ax = plt.subplots(
    1,
    3,
    figsize=(15,5)
)

res = {}

for a, n in zip(ax, nv):

    x = np.random.binomial(
        1,
        0.5,
        size=(m,n)
    )

    s = x.sum(axis=1)

    z = (
        s - n/2
    ) / np.sqrt(n/4)

    mu = np.mean(z)

    var = np.var(z)

    res[n] = (mu,var)

    a.hist(
        z,
        bins=50,
        density=True,
        alpha=0.7,
        label="Sim"
    )

    t = np.linspace(
        -4,
        4,
        300
    )

    a.plot(
        t,
        norm.pdf(t),
        'r-',
        lw=2,
        label="N(0,1)"
    )

    a.set_title(
        f"n={n}\nmean={mu:.4f}, var={var:.4f}"
    )

    a.legend(
        fontsize=8
    )

plt.tight_layout()

plt.savefig(
    "q1_plot.png",
    dpi=150
)

plt.close()

print("Q1 Results")

for n,(mu,var) in res.items():

    print(
        f"n={n}  mean={mu:.4f}  var={var:.4f}"
    )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

np.random.seed(42)

it,_ = quad(
    lambda x: np.exp(-x**2),
    0,
    1
)

print(
    f"True value = {it:.8f}"
)

nv = [
    100,
    1000,
    10000,
    100000
]

est = []

err = []

for n in nv:

    u = np.random.uniform(
        0,
        1,
        n
    )

    ih = np.mean(
        np.exp(-u**2)
    )

    est.append(ih)

    err.append(
        abs(ih-it)
    )

    print(
        f"N={n:6d}  Est={ih:.8f}  Err={abs(ih-it):.2e}"
    )

fig,ax = plt.subplots(
    1,
    2,
    figsize=(12,5)
)

ax[0].semilogx(
    nv,
    est,
    'o-',
    lw=2,
    label='MC'
)

ax[0].axhline(
    it,
    ls='--',
    color='red',
    label=f'{it:.5f}'
)

ax[0].set_xlabel(
    'N'
)

ax[0].set_ylabel(
    'Estimate'
)

ax[0].set_title(
    'Estimate vs N'
)

ax[0].legend()

ax[0].grid(
    alpha=0.3
)

ax[1].loglog(
    nv,
    err,
    'o-',
    lw=2
)

ax[1].set_xlabel(
    'N'
)

ax[1].set_ylabel(
    'Error'
)

ax[1].set_title(
    'Error vs N'
)

ax[1].grid(
    alpha=0.3
)

plt.tight_layout()

plt.savefig(
    "q2_plot.png",
    dpi=150
)

plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

nv = [
    100,
    1000,
    10000,
    100000
]

est = []

for n in nv:

    x = np.random.uniform(
        -1,
        1,
        n
    )

    y = np.random.uniform(
        -1,
        1,
        n
    )

    ins = (
        x**2 + y**2
    ) <= 1

    pi = 4 * np.mean(ins)

    est.append(pi)

    print(
        f"N={n:6d}  pi={pi:.6f}  err={abs(pi-np.pi):.6f}"
    )

# Scatter

m = 5000

x = np.random.uniform(
    -1,
    1,
    m
)

y = np.random.uniform(
    -1,
    1,
    m
)

ins = (
    x**2 + y**2
) <= 1

fig,ax = plt.subplots(
    1,
    2,
    figsize=(13,5)
)

ax[0].scatter(
    x[ins],
    y[ins],
    s=1,
    alpha=0.5,
    label="Inside"
)

ax[0].scatter(
    x[~ins],
    y[~ins],
    s=1,
    alpha=0.5,
    label="Outside"
)

t = np.linspace(
    0,
    2*np.pi,
    300
)

ax[0].plot(
    np.cos(t),
    np.sin(t),
    'k-'
)

ax[0].set_aspect(
    'equal'
)

ax[0].set_title(
    f"N={m}\npi={4*np.mean(ins):.4f}"
)

ax[0].legend(
    fontsize=8
)

# Estimate plot

ax[1].semilogx(
    nv,
    est,
    'o-',
    lw=2,
    label='MC'
)

ax[1].axhline(
    np.pi,
    ls='--',
    color='red',
    label=f'{np.pi:.5f}'
)

ax[1].set_xlabel(
    'N'
)

ax[1].set_ylabel(
    'pi'
)

ax[1].set_title(
    'pi Estimate vs N'
)

ax[1].legend()

ax[1].grid(
    alpha=0.3
)

plt.tight_layout()

plt.savefig(
    "q3_plot.png",
    dpi=150
)

plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

s0 = 100

k = 110

n = 100000

# Main simulation

sig = 0.2

z = np.random.standard_normal(n)

st = s0 * np.exp(sig*z)

pay = np.maximum(
    st-k,
    0
)

ep = np.mean(pay)

print(
    f"sigma=0.20  payoff={ep:.4f}"
)

fig,ax = plt.subplots(
    1,
    2,
    figsize=(13,5)
)

ax[0].hist(
    st,
    bins=100,
    density=True,
    alpha=0.75
)

ax[0].axvline(
    k,
    ls='--',
    color='red',
    label=f'K={k}'
)

ax[0].axvline(
    s0,
    ls='--',
    color='green',
    label=f'S0={s0}'
)

ax[0].set_title(
    'Terminal Prices'
)

ax[0].legend()

ax[1].hist(
    pay,
    bins=80,
    density=True,
    alpha=0.75
)

ax[1].set_title(
    f'Payoffs\nE={ep:.4f}'
)

plt.tight_layout()

plt.savefig(
    "q4_plot1.png",
    dpi=150
)

plt.close()


# Volatility study

sg = [
    0.1,
    0.2,
    0.3,
    0.5
]

col = [
    'steelblue',
    'orange',
    'green',
    'crimson'
]

pm = []

fig,ax = plt.subplots(
    1,
    2,
    figsize=(14,5)
)

for s,c in zip(sg,col):

    z = np.random.standard_normal(n)

    st = s0 * np.exp(s*z)

    pay = np.maximum(
        st-k,
        0
    )

    mp = np.mean(pay)

    pm.append(mp)

    print(
        f"sigma={s:.1f}  payoff={mp:.4f}  std={np.std(st):.4f}"
    )

    ax[0].hist(
        st,
        bins=100,
        density=True,
        alpha=0.5,
        color=c,
        label=f'{s}'
    )

ax[0].axvline(
    k,
    ls='--',
    color='black'
)

ax[0].set_title(
    'ST vs Volatility'
)

ax[0].legend()

ax[1].bar(
    [str(s) for s in sg],
    pm,
    color=col
)

ax[1].set_title(
    'Payoff vs Volatility'
)

for i,v in enumerate(pm):

    ax[1].text(
        i,
        v+0.3,
        f"{v:.2f}",
        ha='center'
    )

plt.tight_layout()

plt.savefig(
    "q4_plot2.png",
    dpi=150
)

plt.close()
```
